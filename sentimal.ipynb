```python
import nltk
from nltk.corpus import stopwords
import spacy

# Download stopwords if not already downloaded
nltk.download('stopwords')

# Load English language model from spaCy
nlp = spacy.load("en_core_web_sm")

# Define stopwords
stop_words = set(stopwords.words("english"))

# Define preprocess function
def preprocess_text(text):
    # Tokenize and lowercase the text using spaCy
    doc = nlp(text.lower())
    
    # Filter out stopwords and non-alphabetic tokens
    tokens = [token.text for token in doc if token.is_alpha and token.text not in stop_words]
    
    # Join tokens back into a single string
    return " ".join(tokens)

# Apply preprocessing function to each review in the dataset
data['processed_review'] = data['reviews'].apply(preprocess_text)
```

```python
%%writefile app.py
import streamlit as st
import spacy
import pickle

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Load trained sentiment analysis model and vectorizer
with open("sentiment_model.pkl", "rb") as f:
    sentiment_model = pickle.load(f)

with open("vectorizer.pkl", "rb") as f:
    vectorizer = pickle.load(f)

# Function to preprocess text
def preprocess_text(text):
    return text.lower()

# Function to extract named entities
def extract_entities(text):
    doc = nlp(text)
    return [(ent.text, ent.label_) for ent in doc.ents]

# Streamlit App
st.title("Sentiment Analysis and NER Pipeline for E-commerce Reviews")

# Input box
review = st.text_area("Enter a review:")

if st.button("Analyze"):
    processed = preprocess_text(review)
    features = vectorizer.transform([processed])
    
    # Predict sentiment
    sentiment = sentiment_model.predict(features)[0]

    # Extract named entities
    entities = extract_entities(review)

    # Show results
    st.subheader("Predicted Sentiment:")
    st.success(sentiment)

    st.subheader("Named Entities:")
    if entities:
        for entity, label in entities:
            st.write(f"- **{entity}** ({label})")
    else:
        st.info("No named entities found.")
```
